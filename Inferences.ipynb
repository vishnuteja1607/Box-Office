```python
#understand budgets
def get_budgets(df):
    
    bins = [0, 50, 100, 150, 200, np.inf]

    # Use pd.cut to categorize values into bins
    categories = pd.cut(df['Budgets'], bins=bins, labels=['$0-50', '$50-100', '$100-150', '$150-200', '$200+'])

    # Group by categories and count the occurrences in each bin
    bin_counts = categories.value_counts().sort_index()
    
    plt.figure(figsize = (12,8))
    
    bars=plt.bar(bin_counts.index, bin_counts.values)
    for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05 * yval, f'{yval}', ha='center', va='bottom')
   
    plt.xlabel("Budgets(in millions USD)")
    plt.ylabel("Number of Releases")
    plt.title("Movies Released v/s Their Budgets, 2000-2019")
    total_things_line = f"Total Releases = {len(df)}"
    plt.text(2.5, max(bin_counts.values) + 5, total_things_line,fontsize=12, ha='center')

    plt.show()
    
get_budgets(df[df['year']==2019])
```

```python
def budgets_per_year(data,year):
    df = data.copy()
    df = df[df['year']==year]
    
    bins = [0, 50, 100, 150, 200, np.inf]

    # Use pd.cut to categorize values into bins
    categories = pd.cut(df['Budgets'], bins=bins, labels=['$0-50', '$50-100', '$100-150', '$150-200', '$200+'])

    # Group by categories and count the occurrences in each bin
    bin_counts = categories.value_counts().sort_index()
    
    plt.figure(figsize = (12,8))
    
    bars=plt.bar(bin_counts.index, bin_counts.values)
    for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05 * yval, f'{yval}', ha='center', va='bottom')
   
    plt.xlabel("Budgets(in millions USD)")
    plt.ylabel("Number of Releases")
    plt.title(f"Movies Released v/s Their Budgets in {year}")
    total_things_line = f"Total Releases = {len(df)}"
    plt.text(3, 0.75*max(bin_counts.values) + 5, total_things_line,fontsize=12, ha='center')

    plt.show()

budgets_per_year(df,2018)
```

```python
def gvb(data):
    
    df = data.copy()
    
    bins = [0, 50, 100, 150, 200, np.inf]

    # Use pd.cut to categorize values into bins
    df['categories'] = pd.cut(df['Budgets'], bins=bins, labels=['$0-50', '$50-100', '$100-150', '$150-200', '$200+'])
    grouped_data = df.groupby('categories')['Total Gross'].sum()
    grouped_data = grouped_data/1e3
    plt.figure(figsize = (12,8))
    
    bars=plt.bar(grouped_data.index, grouped_data.values)
    for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05 * yval, f'${yval:.2f}', ha='center', va='bottom')
   
    plt.xlabel("Budget (in millions) ")
    plt.ylabel("Gross (in billions)")
    plt.title("Gross v/s Budgets, 2000-2019")
    total_things_line = f"Total Gross = ${df['Total Gross'].sum()/1e3:.2f} billion"
    plt.text(2.5, 0.75*max(grouped_data.values) , total_things_line,fontsize=12, ha='center')

    plt.show()
gvb(df)    
```

```python
def gvb_per_year(data,year):
    
    df = data.copy()
    df = df[df['year']==year]
    bins = [0, 50, 100, 150, 200, np.inf]

    # Use pd.cut to categorize values into bins
    df['categories'] = pd.cut(df['Budgets'], bins=bins, labels=['$0-50', '$50-100', '$100-150', '$150-200', '$200+'])
    grouped_data = df.groupby('categories')['Total Gross'].sum()
    grouped_data = grouped_data/1e3
    plt.figure(figsize = (10,8))
    
    bars=plt.bar(grouped_data.index, grouped_data.values)
    for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05 * yval, f'${yval:.2f}', ha='center', va='bottom')
   
    plt.xlabel("Budget (in millions) ")
    plt.ylabel("Gross (in billions)")
    plt.title(f"Gross v/s Budgets during {year}")
    total_things_line = f"Total Gross = ${df['Total Gross'].sum()/1e3:.2f}"
    plt.text(3, 0.75*max(grouped_data.values), total_things_line,fontsize=12, ha='center')

    plt.show()
gvb_per_year(df,2016)
```

```python
def yearly_gb(df):
    
    data = df.copy()
    data = data.groupby('year')
    
    ys1 = data['Total Gross'].sum()/1e3
    ys2 = data['Budgets'].sum()/1e3
    
    xs = np.linspace(2000, 2019, num=20, dtype=int)
    #print(xs)
    plt.figure(figsize=(12,6))
    plt.grid(True)
    plt.plot(xs,ys1,marker='o',label='Gross')
    plt.plot(xs,ys2,marker='o',label='Budgets')
    
    plt.xlabel('Year')
    plt.ylabel('Gross(in Billion USD)')
    plt.title('Total Gross and Budgets per year, 2000-2019')
    plt.xticks(xs)
    plt.legend()
    plt.show()
    
    
yearly_gb(df)
```

```python
#releases vs theaters

def get_theaters(df):
    
    bins = [500,1000,1500,2000,2500,3000,3500,4000,4500,5000]
    
    # Use pd.cut to categorize values into bins
    categories = pd.cut(df['Theaters'], bins=bins, labels=['0-500','500-1000','1000-1500','1500-2000','2000-2500',
                                                         '2500-3000','3000-3500','3500-4000','4000-4500'])

    # Group by categories and count the occurrences in each bin
    bin_counts = categories.value_counts().sort_index()
    plt.figure(figsize = (16,8))
    bars=plt.bar(bin_counts.index, bin_counts.values)
    for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05 * yval, f'{yval}', ha='center', va='bottom')
    plt.xlabel("Theaters")
    plt.ylabel("Number of Releases")
    plt.title("Movies Released v/s Number of Theaters, 2000-2019")
    total_things_line = f"Total Theaters = {len(df)}"
    plt.text(2.5, max(bin_counts.values) + 5, total_things_line,fontsize=12, ha='center')

    plt.show()
get_theaters(df)
```

```python
def get_t_stats(df,t_low,t_high,year):
    
    #df is the main df
    if year is None:
        set1 = df[(df['Theaters']>t_low)&(df['Theaters']<t_high)]
    else:
        data = df[df['year']==year]
        set1 = data[(data['Theaters']>t_low)&(df['Theaters']<t_high)]
    
    fig, axs = plt.subplots(1,2,figsize=(12,8))

    sb.scatterplot(x='Theaters',y='Budgets',data=set1,hue='Potential', legend='full',ax = axs[0])
    sb.scatterplot(x='Theaters',y='Total Gross',data=set1,hue='Potential', legend='full',ax=axs[1])
    axs[0].grid(True)
    axs[1].grid(True)
    plt.show()

    totals = len(set1)
    hits = len(set1[set1['Potential']=='Blockbuster'])
    decents = len(set1[set1['Potential']=='Decent'])
    flops = len(set1[set1['Potential']=='Flop'])
    
    print(f"{totals} movies were released in {t_low} to {t_high} theaters")
    print(f"There were {hits} blockbusters, ratio is {hits/totals:0.2f}")
    print(f"There were {decents} decents, ratio is {decents/totals:0.2f}")
    print(f"There were {flops} flops,ratio is {flops/totals:0.2f}")
get_t_stats(df,1000,4000,None)
```

```python
def gvt_season(df):
    
    set1 = df.copy()
    conditions = [
    (set1['MoR'].isin([5, 6,7])),
    (set1['MoR'].isin([11, 12]))
]
    values = ['Summer', 'Winter']

# Create the new column 'B' based on the conditions
    set1['Season'] = np.select(conditions, values)#, default='Other')
    
    colors = {'Summer': 'red', 'Winter': 'blue'}#, 'Other': 'green'}

    set1['Color'] = set1['Season'].map(colors)
    
    sb.scatterplot(x='Theaters', y='Total Gross', hue='Season', palette=colors, data=set1.dropna(), legend='full')

    
gvt_season(df)
```
